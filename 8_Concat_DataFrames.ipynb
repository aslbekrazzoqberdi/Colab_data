```python
import pandas as pd
```

```python
india_weather = pd.DataFrame({
    "city": ["mumbai", "delhi", "Banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city" : ["new york", "chicago", "orlando"],
    "temperature": [14, 23,25],
    "humidity" : [20, 40,60]

})
us_weather
```

```python
df = pd.concat([india_weather, us_weather])
df
```

```python
df2 = pd.concat([india_weather, us_weather], keys = ["indian", "us"])
df2
```

```python
df2.loc["indian"]
```

```python
df2.loc["us"]
```
