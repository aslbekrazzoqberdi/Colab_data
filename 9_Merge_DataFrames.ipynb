```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    "city": ["new york", "chicago", "orlando", "new jersi"],
    "temperature": [21, 22, 23,55],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["new york", "chicago", "orlando", "vashington"],
    "humidity": [65,68,70,50],
})
df2
```

### how to merge data frames


```python
df3 = pd.merge(df1, df2, on = "city", how = "left")
df3
```

### Merging qilayotganda faqat 2 ta datasetdagi kesishgan qiymatlar qoladi

```python
df4 = pd.merge(df1,df2, on = "city", how = "right")
df4
```

```python
df5 = pd.merge(df1,df2, on = "city", how = "outer", indicator = True)
df5
```
