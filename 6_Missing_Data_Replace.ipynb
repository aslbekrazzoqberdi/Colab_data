```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/py/pandas/6_handling_missing_data_replace/weather_data.csv')
df
```

```python
new_df = df.replace(-99999, np.NaN)
new_df
```

```python
new_df2 = df.replace([-99999, 0], np.NaN)
new_df2
```

```python
new_df3 = df.replace({
    "temperature": -99999,
    "windspeed" : -99999,
    "event" : '0'
}, np.NaN)
new_df3
```
